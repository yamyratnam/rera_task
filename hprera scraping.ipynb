```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options=chrome_options)
```

```python
driver.get("https://hprera.nic.in/PublicDashboard")
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Use WebDriverWait to wait for the element to be present
element = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "tab_project_main-filtered-data"))
)
```

```python
# Locate the div containing the registered projects
registered_projects = driver.find_element(By.ID, "reg-Projects")
```

```python
print(registered_projects)
```

```python
# Find all elements with class 'col-lg-6' under the 'reg-Projects' div
project_elements = registered_projects.find_elements(By.CLASS_NAME, "col-lg-6")
```

```python
print(project_elements)
```

```python
for project in project_elements:
    project_name = project.find_element(By.CLASS_NAME, "font-lg").text
    project_id = project.find_element(By.CSS_SELECTOR, "a[title='View Application']").text
    project_type = project.find_element(By.CSS_SELECTOR, "span:nth-of-type(2)").text
    contact_info = project.find_elements(By.CSS_SELECTOR, "div.mt-1 > span")
    phone = contact_info[0].text
    email = contact_info[1].text
    address = contact_info[2].text
    validity = project.find_element(By.CSS_SELECTOR, "span.text-orange").text

    print(f"Project Name: {project_name}")
    print(f"Project ID: {project_id}")
    print(f"Project Type: {project_type}")
    print(f"Contact Phone: {phone}")
    print(f"Contact Email: {email}")
    print(f"Address: {address}")
    print(f"Validity: {validity}")
    print("\n")
```

```python
# Loop through project elements
for project in project_elements[:5]:  
    project_name = project.find_element(By.CLASS_NAME, "font-lg").text
    project_id_element = project.find_element(By.CSS_SELECTOR, "a[title='View Application']")
    project_id = project_id_element.text
    project_link = project_id_element.get_attribute("href")  # Extracting the link
    
    # Output project details
    print(f"Project Name: {project_name}")
    print(f"Project ID: {project_id}")
    
    # Open the link
    driver.execute_script("arguments[0].click();", project_id_element)

    # Extract information
    # Find the table containing the details
    details_table = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#project-menu-html table.table"))
    )

    # Extract information
    details_rows = details_table.find_elements(By.XPATH, "//table[@class='table table-borderless table-sm table-responsive-lg table-striped font-sm']/tbody/tr")

    details = {}
    for row in details_rows:
        columns = row.find_elements(By.TAG_NAME, "td")
        if len(columns) == 2:
            label = columns[0].text
            value = columns[1].text.strip()
            details[label] = value

    # Accessing specific details
    print("Name:", details.get("Name"))
    print("PAN No.:", details.get("PAN No."))
    print("GSTIN No.:", details.get("GSTIN No."))
    print("Permanent Address:", details.get("Permanent Address"))

    # Find the close button element
    close_button = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Close']"))
    )

    # Click on the close button to close the modal
    close_button.click()
    print("\n")
```

```python
# Loop through project elements
for project in project_elements[:5]:  
    project_name = project.find_element(By.CLASS_NAME, "font-lg").text
    project_id_element = project.find_element(By.CSS_SELECTOR, "a[title='View Application']")
    project_id = project_id_element.text
    project_link = project_id_element.get_attribute("href")  # Extracting the link
    
    # Output project details
    print(f"Project Name: {project_name}")
    print(f"Project ID: {project_id}")
    
    # Open the link
    driver.execute_script("arguments[0].click();", project_id_element)

    # Extract information
    # Find the table containing the details
    details_table = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#project-menu-html table.table"))
    )

    # Extract information
    details_rows = details_table.find_elements(By.XPATH, "//table[@class='table table-borderless table-sm table-responsive-lg table-striped font-sm']/tbody/tr")

    details = {}
    for row in details_rows:
        columns = row.find_elements(By.TAG_NAME, "td")
        if len(columns) == 2:
            label = columns[0].text
            value_element = columns[1]
            
            span_element = value_element.find_elements(By.XPATH, ".//span[@class='fw-600']")
            link_element = value_element.find_elements(By.XPATH, ".//a")
            
            if span_element and link_element:
                # Both span and link present
                values = []
                for span, link in zip(span_element, link_element):
                    value = {
                        'span_text': span.text.strip(),
                        'link_text': link.text.strip(),
                        'link_url': link.get_attribute('href')
                    }
                    values.append(value)
            elif span_element:
                # Only span present
                values = [span.text.strip() for span in span_element]
            elif link_element:
                # Only link present
                values = [{
                    'link_text': link.text.strip(),
                    'link_url': link.get_attribute('href')
                } for link in link_element]
            else:
                # Neither span nor link present
                values = value_element.text.strip()
            
            # Storing the values in the details dictionary
            details[label] = values

    # Accessing specific details
    print("Details:", details)

    # Find the close button element
    close_button = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Close']"))
    )

    # Click on the close button to close the modal
    close_button.click()
    print("\n")
```

```python
for project in project_elements[:5]:
    project_name = project.find_element(By.CLASS_NAME, "font-lg").text
    project_id_element = project.find_element(By.CSS_SELECTOR, "a[title='View Application']")
    project_id = project_id_element.text
    project_link = project_id_element.get_attribute("href")  # Extracting the link

    # Output project details
    print(f"Project Name: {project_name}")
    print(f"Project ID: {project_id}")

    # Open the link
    driver.execute_script("arguments[0].click();", project_id_element)

    # Extract information
    details_table = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#project-menu-html table.table"))
    )

    # Extracting specific information
    details_rows = details_table.find_elements(By.XPATH, "//table[@class='table table-borderless table-sm table-responsive-lg table-striped font-sm']/tbody/tr")

    details = {
        'Name': '',
        'PAN No.': [],
        'Permanent Address': [],
        'GSTIN No.': []
    }

    for row in details_rows:
        columns = row.find_elements(By.TAG_NAME, "td")
        if len(columns) == 2:
            label = columns[0].text
            value_element = columns[1]

            if label == 'Name':
                details['Name'] = value_element.text.strip()

            if label == 'Permanent Address':
                address_elements = value_element.find_elements(By.XPATH, ".//span | .//a")
                address_info = []
                for address_element in address_elements:
                    address_text = address_element.text.strip()
                    address_link = address_element.get_attribute('href') if address_element.tag_name == 'a' else ''
                    address_info.append({'text': address_text, 'link': address_link})
                details['Permanent Address'] = address_info

            if label == 'PAN No.':
                link_elements = value_element.find_elements(By.XPATH, ".//span | .//a")
                links_info = []
                for link_element in link_elements:
                    link_text = link_element.text.strip()
                    link_url = link_element.get_attribute('href') if link_element.tag_name == 'a' else ''
                    links_info.append({'text': link_text, 'link': link_url})
                if links_info:
                    details['PAN No.'] = links_info

            if label == 'GSTIN No.':
                link_elements = value_element.find_elements(By.XPATH, ".//span | .//a")
                links_info = []
                for link_element in link_elements:
                    link_text = link_element.text.strip()
                    link_url = link_element.get_attribute('href') if link_element.tag_name == 'a' else ''
                    links_info.append({'text': link_text, 'link': link_url})
                if links_info:
                    details['GSTIN No.'] = links_info

    # Printing the filtered details
    print("Filtered Details:", details)

    # Closing the modal
    close_button = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Close']"))
    )
    close_button.click()
    print("\n")
```
